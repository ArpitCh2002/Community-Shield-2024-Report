## xG Flowchart Function for Dashboard

```python
h_xG = [0]
a_xG = [0]
h_min = [0]
a_min = [0]

# Finding home and away teams

hteam = df_xg['Team'].iloc[0]
ateam = df_xg['Team'].iloc[-1]

for x in range(len(df_xg['xG'])):
    if df_xg['Team'][x] == ateam:
        a_xG.append(df_xg['xG'][x])
        a_min.append(df_xg['Minute'][x])
    if df_xg['Team'][x] == hteam:
        h_xG.append(df_xg['xG'][x])
        h_min.append(df_xg['Minute'][x])
```

```python
a_xG
```

```python
# Creating a cumulative value function

def nums_cumulative_sum(nums_list):
    return [sum(nums_list[:i+1]) for i in range(len(nums_list))]

a_cumulative = nums_cumulative_sum(a_xG)
h_cumulative = nums_cumulative_sum(h_xG)
```

```python
a_total = round(a_cumulative[-1],2)
h_total = round(h_cumulative[-1],2)
```

```python
# Plotting the xG flowchart for data

fig, ax = plt.subplots(figsize=(10,5))
fig.set_facecolor('#3d4849')
ax.patch.set_facecolor('#3d4849')

plt.xticks([0,15,30,45,60,75,90])
plt.xlabel('Match Minutes', color='#FFFFFF')
plt.ylabel('xG', color='#FFFFFF')

plt.gca().spines['top'].set_color('white')
plt.gca().spines['bottom'].set_color('white')
plt.gca().spines['left'].set_color('white')
plt.gca().spines['right'].set_color('white')

plt.gca().tick_params(axis='x', colors='white')
plt.gca().tick_params(axis='y', colors='white')

ax.step(x=a_min, y=a_cumulative, where='post', color='red', label='Manchester United')
ax.step(x=h_min, y=h_cumulative, where='post', color='skyblue', label='Manchester City')

ax.legend(loc='upper left', facecolor='#3d4849', edgecolor='white', fontsize=10, labelcolor='white')

plt.savefig('Downloads\\xg_chart.png', format='png', dpi=300, bbox_inches='tight', facecolor=fig.get_facecolor())
```

## Plotting Pass Map and Heat Map for Kovacic

```python
from matplotlib.patches import Patch

pitch = Pitch()
# specifying figure size (width, height)
fig, ax = pitch.draw(figsize=(8, 4))

# Creating Pass Map
for x in range(len(df_kova['x'])):
    if df_kova['outcome'][x] == 'Successful':
        plt.plot((df_kova['x'][x], df_kova['endX'][x]) , (df_kova['y'][x], df_kova['endY'][x]) , color='green')
        plt.scatter(df_kova['x'][x] , df_kova['y'][x] , color='green')
    if df_kova['outcome'][x] == 'Unsuccessful':
        plt.plot((df_kova['x'][x], df_kova['endX'][x]) , (df_kova['y'][x], df_kova['endY'][x]) , color='red')
        plt.scatter(df_kova['x'][x] , df_kova['y'][x] , color='red')
        
plt.title('Mateo Kovacic vs Manchester United, Community Shield 2024', color='skyblue', size=11)
        
success_patch = Patch(color='green', label='Successful Pass')
unsuccess_patch = Patch(color='red', label='Unsuccessful Pass')
plt.legend(handles=[success_patch, unsuccess_patch], loc='upper right')

plt.savefig('Downloads\\kovacic_passmap.png', format='png', dpi=300)
```
